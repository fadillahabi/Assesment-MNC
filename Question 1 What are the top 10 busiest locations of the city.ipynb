```python
import pandas as pd
```

```python
data = pd.read_csv("trip_data_4.csv")
```

```python
data.head(100)
```

```python
data.groupby([' pickup_longitude', ' pickup_latitude']).agg({' pickup_datetime':'count'}).sort_values(by=' pickup_datetime', ascending=False).reset_index().head(10)

#[' pickup_datetime', ' pickup_longitude', ' pickup_latitude']
```

```python
data.groupby([' dropoff_longitude', ' dropoff_latitude']).agg({' pickup_datetime':'count'}).sort_values(by=' pickup_datetime', ascending=False).reset_index().head(10)
```
