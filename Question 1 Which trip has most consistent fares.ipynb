```python
import pandas as pd
```

```python
data1 = pd.read_csv("trip_data_4.csv")
data2 = pd.read_csv("trip_fare_4.csv")
```

```python
data1.head()
```

```python
data2.head()
```

```python
listdata = data1.join(data2, lsuffix='_caller', rsuffix='_other')
```

```python
listdata.info()
```

```python
listdata.groupby([' trip_distance', ' pickup_longitude', ' pickup_latitude', ' dropoff_longitude', ' dropoff_latitude']).agg({' total_amount':'mean'}).sort_values(by=' total_amount', ascending=False).head()
```
